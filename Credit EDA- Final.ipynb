#### For Outliers in continoues columns ww eill use IQR method

```python
#### treatment for outlioers in continoues columns ww eill use iqr method
```

```python
# removing outliers using 1.5 IQR method

outl =['AMT_INCOME_TOTAL','AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'EXT_SOURCE_3' ]
for i in outl: 
    q1 = na[i].quantile(0.25)
    q3 = na[i].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    na1 = na.loc[(na[i] > range_low) & (na[i] < range_high)]
```

```python
#### Rechecking outliers

outliers1 =['AMT_INCOME_TOTAL','AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'DAYS_BIRTH','DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH' , 'EXT_SOURCE_2', 'EXT_SOURCE_3' ]
plt.figure(figsize=(6,4))
for i in outliers1:
    print(i)
    sns.boxplot(na1[i], color='red')
    plt.show()
```

##### Creating Bins

###### for AMT_INCOME_TOTAL & AMT_CREDIT,AMT_GOODS_PRICE & DAYS_BIRTH  for more useful study analysis 

```python
# AMT_INCOME_TOTAL- aib

q1=na1['AMT_INCOME_TOTAL'].quantile(0.25)
q2=na1['AMT_INCOME_TOTAL'].quantile(0.50)
q3=na1['AMT_INCOME_TOTAL'].quantile(0.75)
m=na1['AMT_INCOME_TOTAL'].max()

# Binning AMT_INCOME_TOTAL into AMT_INCOME_TOTAL_bin so we don't loose data and have binned values

na1['aib'] = pd.cut(na1['AMT_INCOME_TOTAL'],[q1, q2, q3,m ], labels = ['Low', 'medium', 'High'])
print(na1.aib.value_counts())
```

```python
# AMT_CREDIT- aic

q1=na1['AMT_CREDIT'].quantile(0.25)
q1=na1['AMT_CREDIT'].quantile(0.25)
q2=na1['AMT_CREDIT'].quantile(0.50)
q3=na1['AMT_CREDIT'].quantile(0.75)
m=na1['AMT_CREDIT'].max()

# Binning AMT_CREDIT into AMT_CREDIT_bin so we don't loose data and have binned values

na1['aic'] = pd.cut(na1['AMT_CREDIT'],[q1, q2, q3,m ], labels = ['Low', 'medium', 'High'])
print(na1.aic.value_counts())
```

```python
# AMT_GOODS_PRICE- amg

q1=na1['AMT_GOODS_PRICE'].quantile(0.25)
q2=na1['AMT_GOODS_PRICE'].quantile(0.50)
q3=na1['AMT_GOODS_PRICE'].quantile(0.75)
m=na1['AMT_GOODS_PRICE'].max()

# Binning AMT_CREDIT into AMT_CREDIT_bin so we don't loose data and have binned values

na1['amg'] = pd.cut(na1['AMT_GOODS_PRICE'],[q1, q2, q3,m ], labels = ['Low', 'medium', 'High'])
print(na1.amg.value_counts())
```

###  Outliers Checking and Treatment
#### 
#### we have already cleaned a major part of data and we will keep it so.
#### Hence will create a new data set as pa1 for previoues applicants for outliers, to keep the sanaity of data and use them for analysis as per need
#### now we have 2 data sets for previous applicants after cleaning - pa, and for outliers and further pa1

```python
### outliers

cont=['AMT_ANNUITY','AMT_APPLICATION','AMT_CREDIT','AMT_GOODS_PRICE','CNT_PAYMENT' ]

for i in cont:
    print(i)
    sns.boxplot(pa[i])
    plt.show()
```

##### Using IQR method for outliers treatment

```python
##### Using IQR method for outliers treatment

for i in cont: 
    q1 = pa[i].quantile(0.25)
    q3 = pa[i].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    pa1 = pa.loc[(pa[i] > range_low) & (pa[i] < range_high)]
```

```python
pa1.isnull().sum()
```

### Creating Bins
